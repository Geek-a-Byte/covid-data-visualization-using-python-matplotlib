```python
import pandas as pd
import matplotlib.pyplot as plt
data = pd.read_csv('district.csv')
# https://www.w3resource.com/pandas/dataframe/dataframe-head.php#:~:text=The%20head()%20function%20is,(self%2C%20n%3D5)
# data.head()
# data

# re stores Recovered corona patients count for all districts.
re=data.iloc[:30,5].values
# de stores Deceased corona patients count for all districts.
de=data.iloc[:30,4].values
# co stores Confirmed corona patients count for all districts. 
co=data.iloc[:30,3].values
# x stored District names.
x=list(data.iloc[:30,0])

plt.figure(figsize=(25,10))
ax=plt.axes()
ax.set_facecolor('black')
ax.grid(linewidth=0.4, color='#8f8f8f')
plt.xticks(rotation='vertical',size='20',color='black')#ticks of X - districts
plt.yticks(size='20',color='black')
ax.set_xlabel('\nDistrict',size=25,color='#4bb4f2')
ax.set_ylabel('No. of cases\n',size=25,color='#4bb4f2')
plt.tick_params(size=20,color='black')
ax.set_title('Maharashtra District wise breakdown\n',size=50,color='#28a9ff')

# To plot a bar graph we call .bar and pass it x-axis and y-axis values. 
# Over here we called the plot function three times for all three cases (i.e Recovered , Confirmed, Deceased) 
# all three values are plotted with respect to y-axis and x-axis being common for all three which is district names.
# the sequence of coloring matters, the latter line of code will override the color of previous line of code
plt.bar(x,co,color='#0C1446')
plt.bar(x,re,color='#2B7C85')
plt.bar(x,de,color='#87ACA3')

# we annotate only for Confirmed numbers by iterating over Confirmed cases value.
for i,j in zip(x,co):
    ax.annotate(str(int(j)),xy=(i,j+3),color='white',size='15')

# we have mentioned .legend to indicate legends of the graph. - in the upper right corner
plt.legend(['Confirmed','Recovered','Deceased'],fontsize=20)
```
