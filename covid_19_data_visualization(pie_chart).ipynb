# showing percentage data on pie chart

```python
import matplotlib.pyplot as plt

# So we have created list slices based on which our Pie Chart will be divided
# and the corresponding activities are it’s valued.

slices = [62, 142, 195]
activities = ['Travel', 'Place Visit', 'Unknown']

# cols is for coloring each slices
# We can define colors to assign by passing a list of corresponding colors. 
cols=['#4C8BE2','#00e061','#fe073a']

# To space out each piece of Pie we can pass on the list of corresponding values to ‘explode’.
exp = [0.2,0.02,0.02]

# To plot a Pie Chart we call ‘.pie’ function which takes x values which is ‘slices’ over here 
# based on it the pie is divided followed by labels which have the corresponding string the values it represents. 
# These string values can be altered by ‘textprops’.
# To change the radius or size of Pie we call ‘radius’. 
# For the aesthetics we call ‘shadow’ as True and ‘startangle ’= 90. 

# The ‘autopct’ defines the number of positions that are allowed to be shown. 
# In this case, autopct allows 2 positions before and after the decimal place.

    
plt.pie(slices,labels=activities,
        textprops=dict(size=25,color='black'),
        radius=1.5,
        colors=cols,
        autopct='%2.2f%%',
        explode=exp,
        shadow=True,
        startangle=90)
 
plt.title('Transmission\n\n',color='#4fb4f2',size=40)

# the percentage shown times the sum of all actual values must be the actual value
# (15.54 * 399)/100=62
```

# Showing actual value on pie chart

https://stackoverflow.com/questions/41088236/how-to-have-actual-values-in-matplotlib-pie-chart-displayed

```python
import matplotlib.pyplot as plt
import numpy

slices = numpy.array([62, 142, 195])
activities = ['Travel', 'Place Visit', 'Unknown']
cols=['#4C8BE2','#00e061','#fe073a']
exp = [0.2,0.02,0.02]

def absolute_value(val):
    a  = numpy.round(val/100.*slices.sum(), 0)
    return a
    
plt.pie(slices,labels=activities,
        textprops=dict(size=25,color='black'),
        radius=1.5,
        colors=cols,
        autopct=absolute_value,
        explode=exp,
        shadow=True,
        startangle=90)
 
plt.title('Transmission\n\n',color='#4fb4f2',size=40)
```
